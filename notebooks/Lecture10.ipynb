## Introduction to statistical distributions

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

### Uniform distribution

```python
s = np.random.uniform(0,1,10000)
h = plt.hist(s,20)
```

### One dimensional Gaussian distribution

```python
s = np.random.normal(0,1,10000)
h = plt.hist(s,100)
```

### Two dimensional Gaussian distribution

```python
fig = plt.figure()

s = np.random.multivariate_normal([0,0],[[1,0],[0,1]],10000)

ax = fig.add_subplot(121)
hist,xedges,yedges = np.histogram2d(s[:,0],s[:,1],20)
im = ax.imshow(hist, interpolation='nearest', origin='low',
                extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]])

ax = fig.add_subplot(122, projection='3d')
elements = (len(xedges) - 1) * (len(yedges) - 1)
xpos, ypos = np.meshgrid(xedges[:-1]+0.25, yedges[:-1]+0.25)

xpos = xpos.flatten()
ypos = ypos.flatten()
zpos = np.zeros(elements)
dx = 0.5 * np.ones_like(zpos)
dy = dx.copy()
dz = hist.flatten()

ax.bar3d(xpos, ypos, zpos, dx, dy, dz, color='y', zsort='average')
```
